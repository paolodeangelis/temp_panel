```python
import uuid

import numpy as np
import pandas as pd
```

```python
N_ROW = 10000
```

```python
data_1 = pd.DataFrame(
    {
        "ID": [str(uuid.uuid4())[-12:] for i in range(N_ROW)],
        "Category 1": np.random.choice(["A", "B", "C", "D", "E", "F"], size=N_ROW),
        "Property 1": np.random.normal(loc=10.0, scale=100.0, size=N_ROW),
        "Property 2": np.random.normal(loc=0.0, scale=10.0, size=N_ROW),
        "Property 3": np.exp(
            np.random.normal(loc=3.0, scale=1.2, size=N_ROW)
            + np.random.normal(loc=0.0, scale=0.5, size=N_ROW)
        ),
        "Property 4": np.random.normal(loc=500.0, scale=300.0, size=N_ROW),
        "Property 5": np.random.normal(loc=9000.0, scale=9000.0, size=N_ROW),
        "Property 6": np.random.normal(loc=10.0, scale=100.0, size=N_ROW) ** 4
        + np.random.normal(loc=100.0, scale=500.0, size=N_ROW),
        "Property 7": np.random.normal(loc=0.0, scale=150.0, size=N_ROW)
        + np.random.normal(loc=0.0, scale=10.0, size=N_ROW),
        "Property 8": np.random.normal(loc=0.0, scale=1000.0, size=N_ROW),
    }
)

data_1.to_json("data/mclass1.json")
```

```python
data_2 = pd.DataFrame(
    {
        "ID": [str(uuid.uuid4())[-12:] for i in range(N_ROW)],
        "Category 1": np.random.choice(["A", "B", "C", "D", "E", "F"], size=N_ROW),
        "Property 1": np.random.normal(loc=10.0, scale=1000.0, size=N_ROW),
        "Property 2": np.random.normal(loc=-10.0, scale=1.0, size=N_ROW),
        "Property 3": np.exp(
            np.random.normal(loc=4.0, scale=0.6, size=N_ROW)
            + np.random.normal(loc=0.0, scale=0.5, size=N_ROW)
        ),
        "Property 4": np.random.normal(loc=55.0, scale=76.0, size=N_ROW),
        "Property 5": np.random.normal(loc=123.0, scale=8741.0, size=N_ROW),
        "Property 6": np.random.normal(loc=10.0, scale=100.0, size=N_ROW) ** 4
        + np.random.normal(loc=100.0, scale=500.0, size=N_ROW),
        "Property 7": np.random.normal(loc=0.0, scale=150.0, size=N_ROW)
        + np.random.normal(loc=0.0, scale=10.0, size=N_ROW),
        "Property 8": np.random.normal(loc=-200.0, scale=1000.0, size=N_ROW),
    }
)

data_2.to_json("data/mclass2.json")
```
